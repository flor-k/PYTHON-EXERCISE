```python
import requests
import csv
from bs4 import BeautifulSoup

f = csv.writer(open('prueba8.csv', 'w', newline=''))
f.writerow(['Name', 'Link', 'Datos'])

pages = []

for i in range(1, 5):
    url = 'https://web.archive.org/web/20121007172955/https://www.nga.gov/collection/anZ' + str(i) + '.htm'
    pages.append(url)
    
for item in pages:
    page = requests.get(item)
    soup = BeautifulSoup(page.text, 'html.parser')
    
    last_links = soup.find(class_='AlphaNav')
    last_links.decompose()

    artist_name_list = soup.find(class_='BodyText')
    artist_name_list_items = artist_name_list.find_all('tr')

    for artist_row in artist_name_list_items:
        artist_row_tds = artist_row.find_all('td')
        ahref = artist_row_tds[0].find_all('a')
        href = 'https://web.archive.org' + ahref[0].get('href')
        name = ahref[0].get_text()
        datos = artist_row_tds[1].text
        print(name, href, datos)
        f.writerow([name, href, datos])
```
